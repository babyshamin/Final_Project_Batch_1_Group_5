chi_square test

    to check whether the categorical columns are correlated or not.

    H0: Variables are not corelated.

    This is the null Hypothesis.

```python
from scipy.stats import chi2_contingency
def chisquare(a,b):
  dataset_table =pd.crosstab(cat_cols[a],cat_cols[b])
  stat, p, dof, expected = chi2_contingency(dataset_table)
  print("Stat=",stat,'\n')
  print("p value of",a,"and",b,"=",p,'\n')
  print("Degree of Freedom=",dof,'\n')
  
  # interpret p-value
  alpha = 0.05
  if p <= alpha:
      print('Dependent- shows variables are correlated (reject H0)')
  else:
      print('Independent- shows variables are not correlated (H0 holds true)')
```
